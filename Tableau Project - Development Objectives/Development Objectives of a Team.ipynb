Tableau Project - Analysis on Development Objectives

A Tableau dashboard that analyzes a fictional group's top development needs and activities.

Highlights: 

1) Cleaned and pre-processed data stored in Excel spreadsheet.

2) Used Python to create code that categorizes users' open-ended inputs by keywords and appends them into a new Activity Category column in the spreadsheet.

3) Visualized this information with the help of Tableau.
