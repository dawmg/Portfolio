Tableau Project - Insights on Cathay Pacific's traffic data

All data was collected from publicly available information - Cathay Pacific's annual report 2024 and its monthly traffic data publications.
Example report: https://news.cathaypacific.com/the-cathay-group-releases-traffic-figures-for-december-2025

Process:
1. Exporting the data from the article into a CSV
2. Cleaning data as needed using Python
3. Conducting data visualization via Tableau

Link to project: https://public.tableau.com/app/profile/wenzell.guadiz/viz/InsightsonCathaytrafficdata/INFO
